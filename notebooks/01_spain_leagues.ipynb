```python
import requests
url = "https://server1.api-football.com/leagues/country/Spain"
headers = {'x-rapidapi-key': "28d61524ab5698a1f8edb9e9ed577221"}

response = requests.request("GET", url, headers=headers)
```

```python
spain_leagues = response.json()
spain_leagues
```

```python
league_id = []
season =[]
season_start = []
season_end = []
name = []

for i in spain_leagues['api']['leagues']:
    league_id.append(i['league_id'])
    season.append(i['season'])
    season_start.append(i['season_start'])
    season_end.append(i['season_end'])
    name.append(i['name'])

import pandas as pd
df = pd.DataFrame(list(zip(name, league_id, season, season_start, season_end)), 
                  columns =['name', 'league_id', 'season', 'season_start', 'season_end'])
```

```python
df.sort_values(by=['name', 'season'])
```

```python
df.to_csv("../bbdd/spain_leagues.csv", index=False)
```
