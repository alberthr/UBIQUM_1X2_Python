```python
from itertools import product
import pandas as pd

def expand_grid(dictionary):
    return pd.DataFrame([row for row in product(*dictionary.values())], columns=dictionary.keys())

dictionary = {'p1': ['1', 'X', '2'], 'p2': ['1', 'X', '2'], 'p3': ['1', 'X', '2'], 'p4': ['1', 'X', '2'], 
              'p5': ['1', 'X', '2'], 'p6': ['1', 'X', '2'], 'p7': ['1', 'X', '2'], 'p8': ['1', 'X', '2'], 
              'p9': ['1', 'X', '2'], 'p10': ['1', 'X', '2'], 'p11': ['1', 'X', '2'], 'p12': ['1', 'X', '2'], 
              'p13': ['1', 'X', '2'], 'p14': ['1', 'X', '2']}
```

```python
full_grid = expand_grid(dictionary)
```

```python
from collections import Counter
full_grid['n1'] = full_grid.apply(lambda x: Counter(x[:13])['1'], axis=1)
full_grid['nx'] = full_grid.apply(lambda x: Counter(x[:13])['X'], axis=1)
full_grid['n2'] = full_grid.apply(lambda x: Counter(x[:13])['2'], axis=1)
```
