```python
import requests
url = "https://server1.api-football.com/fixtures/league/775/2019-10-20"
headers = {'x-rapidapi-key': "28d61524ab5698a1f8edb9e9ed577221"}

response = requests.request("GET", url, headers=headers)
```

```python
results = response.json()
```

```python
hometeam = []
awayteam = []
referee = []
goalshome = []
goalsaway = []

for i in results['api']['fixtures']:
    hometeam.append(i['homeTeam']['team_name'])
    awayteam.append(i['awayTeam']['team_name'])
    referee.append(i['referee'])
    goalshome.append(i['goalsHomeTeam'])
    goalsaway.append(i['goalsAwayTeam'])
```

```python
goalsaway
```

```python
import pandas as pd

df = pd.DataFrame(list(zip(hometeam, awayteam, referee, goalshome, goalsaway)), 
               columns =['Local', 'Visitor', 'Referee', 'Goals_H', 'Goals_V'])

df
```
